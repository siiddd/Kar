```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_excel(r'https://github.com/kkk575/data.cvv/blob/1698d5650bbcf2bbe17d87918a407edad9df6a35/data%20ccv.xlsx?raw=true')
```

```python
df.head(3)
```

```python
df_nodistricts = df.iloc[:-1, 1:]
df_nodistricts
```

```python
x_array = np.array(df_nodistricts)
```

```python
d = preprocessing.normalize(x_array.data, axis=0)
normalized_df = pd.DataFrame(d, columns=df_nodistricts.columns)
normalized_df
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(d)

Eigen_Vectors = pca.components_
Eigen_Values = pca.explained_variance_
```
